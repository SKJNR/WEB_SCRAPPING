```python
#Importing Necessary Libraries 
from selenium import webdriver
import os 
import shutil
import os
import requests
import pandas as pd
import csv
```

```python
# creating an instance of google chrome
DRIVER_PATH ='F:\data sets\chromedriver.exe.exe'
```

```python
# to run chrome in a headfull moe(like regular chrome)
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
my_page = driver.get("https://images.google.com/")
```

```python
search_bar = driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input') # Finding the search bar using it's xpath

search_bar.send_keys("Flowers")                                            # Inputing "Rock" keyword to search rock images

search_button = driver.find_element_by_xpath('//*[@id="sbtc"]/button') # Finding the xpath of search button

search_button.click()
for _ in range(500):

    driver.execute_script("window.scrollBy(0,10000)")
```

```python
def make_directory(dirname):
    current_path = os.getcwd()
    path = os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
make_directory('flowers')
```

```python
images=driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
```

```python
images[:10]
```

```python
img_urls = []

for image in images:
    source= image.get_attribute('src')
    if source is not None:
        if(source[0:4] == 'http'):
            img_urls.append(source)
img_urls
```

```python
page=1
for index,link in enumerate(img_urls):
    print("Downloading {0} of {1} images from page 1".format(index+1,len(img_urls)))
    response=requests.get(link)
    with open('flowers/img_{0}{1}.jpeg'.format(index,page),"wb") as file:
            file.write(response.content)
```
